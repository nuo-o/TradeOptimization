impute strategy:
1. the forecast is per hour, then evenly shared the power output for the four quarters. 
2. if the hourly-forecast is missing, find next or previous year's data in the same period same day
3. otherwise, missing values are filled with average power output in the same period of the whole data. 

Note: there are about 89% of missingd data in the forecasting data

```python
# get solar and wind value divide by four
solar_wind_col = ['solar_value', 'solar_maxvalue', 'solar_minvalue', 'wind_value', 'wind_maxvalue', 'wind_minvalue']
solar_wind_df = pd.read_excel(weather_folder + '/complete_solar_wind.xlsx')
solar_wind_df[ solar_wind_col ]/=4

solar_wind_df.to_excel(weather_folder + '/averaged_solar_wind.xlsx', index = False)
```

```python
impute_solar_wind_df = pd.read_excel(weather_folder + '/averaged_solar_wind.xlsx')
impute_solar_wind_df.head()
```

```python
import time
start = time.time()

# impute 
for index, row in impute_solar_wind_df.iterrows():
       
    current_PTE = impute_solar_wind_df.loc[index]['PTE']
    
    if row.isnull().any():
        # contain missing values
        if current_PTE % 4 == 1 | impute_solar_wind_df.loc[index-1].isnull().any():
            # missing forecast data
            continue
        
        # impute the same value to four PTE within an hour
        impute_solar_wind_df.loc[index,solar_wind_col] = impute_solar_wind_df.loc[index-1, solar_wind_col]

end = time.time()

print(end - start)
```

```python
impute_solar_wind_df = pd.read_excel(weather_folder + '/averaged_solar_wind.xlsx')

cur_index = 0
import time
start = time.time()

while cur_index < len(impute_solar_wind_df):
#     if cur_index > 20:
#         break
    
    PTE = impute_solar_wind_df.loc[cur_index]['PTE']
#     print(PTE)

    cur_row = impute_solar_wind_df.loc[cur_index][solar_wind_col]
    
    if PTE % 4 == 1 & cur_row.isnull().any():
        cur_index +=96
        continue
    
    if PTE % 4 ==1:
#         print('impute the following three lines')
        impute_solar_wind_df.loc[cur_index+1, solar_wind_col] = cur_row
        impute_solar_wind_df.loc[cur_index+2, solar_wind_col] = cur_row
        impute_solar_wind_df.loc[cur_index+3, solar_wind_col] = cur_row
        cur_index += 4
        continue
    

    raise ValueError('Invalid missing values')
    
end = time.time()

print('takes:{} min'.format((end - start)/60))
impute_solar_wind_df.tail()
```
